```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
results_dir = "../../results/eight/"
```

```python
def plot_hist_output(epoch):
    vals = np.load(results_dir + "epoch_{}_outputs_noise.npy".format(epoch))
    print(vals.shape, vals.mean())
    plt.figure()
    x = np.linspace(-1.,1.,750)
    plt.hist(vals, range=[-1.,1.], bins=150)
    plt.grid(True)
    
def plot_hist_hidden(epoch):
    vals = np.load(results_dir + "epoch_{}_h1_noise.npy".format(epoch))
    print(vals.shape, vals.mean())
    plt.figure()
    x = np.linspace(-1.,5.,750)
    plt.hist(vals, range=[-2.5,5.], bins=150)
    plt.grid(True)

def plot_hist_over_epochs(epochs):
    num_bins = 100
    for epoch in epochs:
        vals = np.load(results_dir + "epoch_{}_outputs_noise.npy".format(epoch))
        hist, _ = np.histogram(vals, range=[-1.,1.], bins=num_bins)
        xmin = np.min(vals)
        xmax = np.max(vals)
        x = np.linspace(xmin, xmax, num_bins)
        plt.plot(x, hist, label="Epoch {}".format(epoch))
        plt.xlim(-1.,1.)
        plt.legend()
    plt.grid(True)

def plot_hist_top_view_hidden(num_epochs):
    num_bins = 100
    points = 1000
    plt.figure(figsize=(10,5))
    for epoch in range(num_epochs):
        vals = np.load(results_dir + "epoch_{}_h1_noise.npy".format(epoch+1))
        hist, bins = np.histogram(vals, range=[-2.5,5.], bins=num_bins)
        total_points = np.sum(hist)
#         assert total_points == points, "Error:{}".format(total_points)
        
        i = 0
        colours = list()
        for b in range(bins.shape[0]-1):
            c = plt.cm.jet(hist[i]/float(total_points)*30)
            colours.append(c)
            i += 1
            
        epoch_vals = np.tile(epoch, bins.shape)
        plt.scatter(epoch_vals, bins, color=colours)
    plt.show()

def plot_hist_top_view_output(num_epochs):
    num_bins = 100
    points = 1000
    plt.figure(figsize=(10,5))
    for epoch in range(num_epochs):
        vals = np.load(results_dir + "epoch_{}_outputs_noise.npy".format(epoch+1))
        hist, bins = np.histogram(vals, range=[-1.,1.], bins=num_bins)
        total_points = np.sum(hist)
#         assert total_points == points, "Error:{}".format(total_points)
        
        i = 0
        colours = list()
        for b in range(bins.shape[0]-1):
            c = plt.cm.jet(hist[i]/float(total_points)*30)
            colours.append(c)
            i += 1
            
        epoch_vals = np.tile(epoch, bins.shape)
        plt.scatter(epoch_vals, bins, color=colours)
    plt.show()

def plot_acc():
    vals = np.load(results_dir + "accuracies.npy")
    plt.figure(figsize=(10,5))
    plt.plot(vals)
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.grid(True)
    
def plot_mutual_info():
    vals = np.load(results_dir + "mutual_information.npy")
    plt.figure(figsize=(10,5))
    plt.plot(vals)
    plt.ylabel("Mutual Information")
    plt.xlabel("Epoch")
    plt.grid(True)
```

```python
plot_hist_hidden(1)
```

```python
plot_hist_output(1)
```

```python
plot_hist_output(500)
```

```python
plot_hist_top_view_hidden(500)
```

```python
plot_hist_top_view_output(500)
```
